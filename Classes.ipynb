````python
#Exercicio 1
class Bola:
  def __init__(self, cor, circunferencia, material):
    self.cor = cor
    self.circunferencia = circunferencia
    self.material = material

  def trocaCor(self, nova_cor):
    self.cor = nova_cor

  def mostraCor(self):
    return self.cor

bola1 = Bola('Azul', 30, 'Couro')
print(f'Cor inicial da bola: {bola1.mostraCor()}')

bola1.trocaCor('Vermelha')
print(f'Novo cor da bola: {bola1.mostraCor()}')
````

````python
#Exercicio 2
class Quadrado:
  def __init__(self, lado):
    self.lado = lado

  def trocaLado(self, novo_lado):
    self.lado = novo_lado

  def mostraLado(self):
    return self.lado

  def calculaArea(self):
    return self.lado ** 2

quadrado = Quadrado(5)
print(f'Tamanho do lado do quadrado: {quadrado.mostraLado()}')

quadrado.trocaLado(10)
print(f'Novo tamanho do lado do quadrado: {quadrado.mostraLado()}')
print(f'Área do quadrado: {quadrado.calculaArea()}')
````

````python
#Exercicio 3
class Retangulo:
  def __init__(self, compimento, largura):
    self.comprimento = compimento
    self.largura = largura

  def trocaLados(self, novo_comprimento, nova_largura):
    self.comprimento = novo_comprimento
    self.largura = nova_largura

  def mostraLados(self):
    return self.comprimento, self.largura

  def area(self):
    return self.comprimento * self.largura

  def perimetro(self):
    return 2 * (self.comprimento + self.largura)

def main():
  print('Informe as medidas do local:')
  comprimento = float(input('Comprimento (em metros): '))
  largura = float(input('Largura (em metros): '))

  retangulo = Retangulo(comprimento, largura)

  print('\nMedidas do local:')
  print(f'Comprimento: {retangulo.mostraLados()[0]} metros.')
  print(f'Largura: {retangulo.mostraLados()[1]} metros.')

  area_local = retangulo.area()
  perimetro_local = retangulo.perimetro()

  print(f'\nÁrea do local: {area_local:.2f} m²')
  print(f'Perímetro do local: {perimetro_local:.2f} metros.')

  tamanho_piso = float(input('\nInforme o tamanho de cada piso (em m²): '))
  pisos_necessarios = area_local / tamanho_piso

  rodape_altura = float(input('Informe a altura do rodapé (em metros): '))
  rodape_necessario = perimetro_local / rodape_altura

  print(f'\nQuantidade de pisos necessários: {pisos_necessarios:.2f}')
  print(f'Quantidade de rodapés necessários: {rodape_necessario:.2f}')

if __name__ == "__main__":
  main()
````

````python
#Exercicio 4
class Pessoa:
  def __init__(self, nome, idade, peso, altura):
    self.nome = nome
    self.idade = idade
    self.peso = peso
    self.altura = altura

  def envelhecer(self, anos):
    for _ in range(anos):
      if anos < 21:
        self.crescer(0.005)
      self.idade +=1

  def engordar(self, quilos):
    self.peso += quilos

  def emagrecer(self, emagrecer):
    self.peso -= emagrecer
    if self.peso < 0:
      self.peso = 0

  def crescer(self, metros):
    self.altura += metros

  def __str__(self):
    return (f"""
Nome: {self.nome}.
Idade: {self.idade} anos.
Peso: {self.peso:.2f} kg.
Altura: {self.altura:.2f} m.""")

if __name__ == "__main__":
  pessoa = Pessoa(nome='Kalleu', idade=22, peso=86, altura=1.91)
  print('Dados inicias da pessoa: ')
  print(pessoa)

  pessoa.envelhecer(5)
  print('\nApós envelhecer 5 anos: ', end='')
  print(pessoa)

  pessoa.engordar(10)
  print('\nApós engordar 10 kg: ', end='')
  print(pessoa)

  pessoa.emagrecer(5)
  print('\nApós emagrecer 5 kg: ', end='')
  print(pessoa)

  pessoa.crescer(0.02)
  print('\nApós crescer 2 cm: ', end='')
  print(pessoa)
````

````python
#Exercicio 5
class ContaCorrente:
  def __init__(self, numero_conta, nome_correntista, saldo=0):
    self.numero_conta = numero_conta
    self.nome_correntista = nome_correntista
    self.saldo = saldo

  def alterarNome(self, alterar_nome):
    self.nome_correntista = alterar_nome

  def deposito(self, valor):
    if valor > 0:
      self.saldo += valor
    else:
      print('O valor do depósito deve ser positivo.')

  def saque(self, valor):
    if valor > 0:
      if self.saldo >= valor:
        self.saldo -= valor
      else:
        print('Saldo insuficiente para o saque.')
    else:
      print('O valor do saque deve ser positivo.')

  def __str__(self):
    return (f"""Nº da Conta: {self.numero_conta}.
Nome: {self.nome_correntista}.
Saldo: R${self.saldo:.2f}.""")

if __name__ == "__main__":
  correntista = ContaCorrente(numero_conta=1, nome_correntista="Kalleu", saldo=500)
  print('Dados do Correntista: ')
  print(correntista)

  correntista.alterarNome("Claudio")
  print('\nNovo nome do Correntista: ')
  print(correntista)

  correntista.deposito(100)
  print('\nValor do deposito: ')
  print(f'R$ {correntista}')

  correntista.saque(300)
  print('\nValor do saque: ')
  print(f'R$ {correntista}')

  correntista.saque(500)
  print('\nTentativa de saque maior que o saldo: ')
  print(f'R$ {correntista}')
````

````python
#Exercicio 6
class TV:
  def __init__(self, canal=1, volume=10):
    self.canal = canal
    self.volume = volume

  def trocarCanal(self, novo_canal):
    if 1 <= novo_canal <= 100:
      self.canal = novo_canal
    else:
      print('Canal inválido. Informe um canal entre 1 e 100.')

  def aumentarVolume(self):
    if self.volume < 100:
      self.volume += 1
    else:
      print('O volume já está no máximo.')

  def diminuirVolume(self):
    if self.volume > 0:
      self.volume -= 1
    elif self.volume == 0:
      print('O volume já está no mínimo.')

  def __str__(self):
    return f'Canal atual: {self.canal}\nVolume atual: {self.volume}'

if __name__ == '__main__':
  minha_TV = TV()
  print('Estado inicial da TV:')
  print(minha_TV)

  print('\nAlterando o canal para 10:')
  minha_TV.trocarCanal(10)
  print(minha_TV)

  print('\nAumento o volume 5 vezes:')
  for _ in range(5):
    minha_TV.aumentarVolume()
  print(minha_TV)

  print('\nDiminuindo o volume 3 vezes.')
  for _ in range(3):
    minha_TV.diminuirVolume()
  print(minha_TV)

  print('\nTentando mudar para um canal inválido (200):')
  minha_TV.trocarCanal(200)
  print(minha_TV)

  print('\nTentando diminuir o voluma até o mínimo:')
  while minha_TV.volume > 0:
    minha_TV.diminuirVolume()
  print(minha_TV)
````

````python
#Exercicio 7
from time import sleep

class Tamagushi:
  def __init__(self, nome, fome, saude, idade):
    self.nome = nome
    self.fome = fome
    self.saude = saude
    self.idade = idade

  def alterarNome(self, novo_nome):
    self.nome = novo_nome

  def alterarFome(self, nova_fome):
    if 0 <= nova_fome <= 100:
      self.fome = nova_fome
    else:
      print('Fome deve estar entre 0 e 100.')

  def alterarSaude(self, nova_saude):
    if 0 <= nova_saude <= 100:
      self.saude = nova_saude
    else:
      print('Saúde deve estar entre 0 e 100.')

  def alterarIdade(self, nova_idade):
    if nova_idade >= 0:
      self.idade = nova_idade
    else:
      print('Idade não pode ser negativa.')

  def calcularHumor(self):
    return (self.saude - self.fome) / 100

  def __str__(self):
    humor = self.calcularHumor()
    return f'Nome: {self.nome}\nFome: {self.fome}\nSaude: {self.saude}\nIdade: {self.idade}\nHumor: {humor:.2f}'

if __name__ == '__main__':
  bichinho = Tamagushi(nome="Jorgin", fome=50, saude=50, idade=5)
  print(f'Estado inicial do Tamagushi: ')
  print(bichinho)

  print('\nAlterando os atributos do Tamagushi...')
  sleep(2.5)
  bichinho.alterarNome('Fofucho')
  bichinho.alterarFome(20)
  bichinho.alterarSaude(90)
  bichinho.alterarIdade(6)
  print('\nEstamos atualizando os dados do Tamagushi...')
  sleep(2.5)
  print(bichinho)
````

````python
#Exercicio 8
from time import sleep

class Macaco:
  def __init__(self, nome):
    self.nome = nome
    self.bucho = []

  def comer(self, alimento):
    if isinstance(alimento, Macaco) and alimento.nome != self.nome:
      print(f'{self.nome} comeu {alimento.nome} 😱!')
      self.bucho.append(alimento.nome)
    elif isinstance(alimento, Macaco) and alimento.nome == self.nome:
      print(f'{self.nome} não pode comer a si mesmo!')
    else:
      print(f'{self.nome} comeu {alimento}')
      self.bucho.append(alimento)

  def verBucho(self):
    if self.bucho:
      print(f"O bucho de {self.nome} contém: {', '.join(str(item) for item in self.bucho)}")
    else:
      print(f'O bucho de {self.nome} está vazio.')

  def digerir(self):
    if self.bucho:
      digerido = self.bucho.pop(0)
      print(f'{self.nome} digeriu {digerido}')
    else:
      print(f'{self.nome} não tem nada no bucho para digerir.')

if __name__ == '__main__':
  macaco1 = Macaco('Jorge')
  macaco2 = Macaco('Jaguarita')

  print('\nAlimentando os macacos:')
  sleep(1)
  macaco1.comer('Banana')
  macaco1.verBucho()
  macaco1.comer('Maçã')
  macaco1.verBucho()

  macaco2.comer('Laranja')
  macaco2.comer('Melancia')
  macaco2.verBucho()

  sleep(1)
  print('\nExperimentando o canibalismo:')
  sleep(1)
  macaco1.comer(macaco2)
  macaco1.verBucho()

  sleep(1)
  print('\nDigerindo os alimentos...')
  sleep(2.5)
  macaco1.digerir()
  macaco1.verBucho()
  macaco1.digerir()
  macaco1.verBucho()
````

````python
#Exercicio 9
class Ponto:
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y

  def __str__(self):
    return f'Ponto(x={self.x}, y={self.y})'

class Retangulo:
  def __init__(self, largura, altura, vertice_inferior_esquerdo):
    self.largura = largura
    self.altura = altura
    self.vertice_inferior_esquerdo = vertice_inferior_esquerdo

  def encontrar_centro(self):
    x_centro = self.vertice_inferior_esquerdo.x + self.largura / 2
    y_centro = self.vertice_inferior_esquerdo.y + self.altura / 2
    return Ponto(x_centro, y_centro)

  def __str__(self):
    return (f'Retângulo(largura={self.largura}, altura={self.altura}, '
            f'vértice_inferior_esquerdo={self.vertice_inferior_esquerdo})')

def menu():
  retangulos = []

  while True:
    print('\nMenu:')
    print('1. Criar novo retângulo')
    print('2. Listar retângulos')
    print('3. Alterar retângulo')
    print('4. Mostrar centro de um retângulo')
    print('5. Sair')
    escolha = input('Escolha uma opção: ')

    if escolha == '1':
      largura = float(input('Informe a largura do retângulo: '))
      altura = float(input('Informe a altura do retângulo: '))
      x = float(input('Informe o valor de x do vértice inferior esquerdo: '))
      y = float(input('Informe o valor de y do vértice inferior esquerdo: '))
      retangulo = Retangulo(largura, altura, Ponto(x, y))
      retangulos.append(retangulo)
      print('Retângulo criado com sucesso!')

    elif escolha == '2':
      if retangulos:
        print('\nLista de Retângulos: ')
        for i, ret in enumerate(retangulos):
          print(f'{i + 1}. {ret}')
      else:
        print('Nenhum retângulo criado ainda.')

    elif escolha == '3':
      if retangulos:
        for i, ret in enumerate(retangulos):
          print(f'{i + 1}. {ret}')
        index = int(input('Escolha o número do retângulo que deseja alterar: ')) - 1

        if 0 <= index < len(retangulos):
          largura = float(input('Informe a nova largura do retângulo: '))
          altura = float(input('Informe a nova altura do retângulo: '))
          x = float(input('Informe o novo valor de x do vértice inferior esquerdo: '))
          y = float(input('Informe o novo valor de y do vértice inferior esquerdo: '))
          retangulos[index] = Retangulo(largura, altura, Ponto(x, y))
          print('Retângulo alterado com sucesso!')
        else:
          print('Número inválido!')
      else:
        print('Nenhum retângulo criado ainda.')

    elif escolha == '4':
      if retangulos:
        for i, ret in enumerate(retangulos):
          print(f'{i + 1}. {ret}')

        try:
          index = int(input('Escolha o número do retângulo para calcular o centro: ')) - 1
          if 0 <= index < len(retangulos):
            centro = retangulos[index].encontrar_centro()
            print(f'O centro do retângulo é: {centro}')
          else:
            print('Número inválido!')
        except ValueError:
          print('Entrada inválida! Digite um número.')
      else:
        print('Nenhum retângulo criado ainda.')

    elif escolha == '5':
      print('Encerrando programa...')
      break
    else:
      print('Opção inválida! Tente novamente.')

if __name__ == '__main__':
  menu()
````

````python
#Exercicio 10
class BombaCombustivel:
  def __init__(self, tipo_combustivel, valor_litro, quantidade_combustivel):
    self.tipo_combustivel = tipo_combustivel
    self.valor_litro = valor_litro
    self.quantidade_combustivel = quantidade_combustivel

  def abastecer_por_valor(self, valor):
    litros_abastecidos = valor / self.valor_litro
    if litros_abastecidos <= self.quantidade_combustivel:
      self.quantidade_combustivel -= litros_abastecidos
      print(f'Abastecido {litros_abastecidos:.2f} litros de {self.tipo_combustivel}.')
    else:
      print('Não há combustível suficente na bomba para este valor.')
      print(f'Quantidade disponível: {self.quantidade_combustivel:.2f} litros.')

  def abastecer_por_litro(self, litros):
    if litros <= self.quantidade_combustivel:
      valor_total = litros * self.valor_litro
      self.quantidade_combustivel -= litros
      print(f'Você abasteceu {litros:.2f} litros de {self.tipo_combustivel} por R${valor_total:.2f}')
    else:
      print('Não há combustível suficente na bomba para este valor.')
      print(f'Quantidade disponível: {self.quantidade_combustivel:.2f} litros.')

  def alterar_valor(self, novo_valor):
    self.valor_litro = novo_valor
    print(f'O valor do litro {self.tipo_combustivel} foi alterado para R${novo_valor:.2f}.')

  def alterar_combustivel(self, novo_combustivel):
    self.tipo_combustivel = novo_combustivel
    print(f'O tipo de combustível foi alterado para {self.tipo_combustivel}.')

  def alterar_quantidade_combustivel(self, nova_quantidade):
    self.quantidade_combustivel = float(nova_quantidade)
    print(f'A quantidade de combustível foi aleterado para {self.quantidade_combustivel:.2f} litros.')

  def __str__(self):
    return (f'Tipo de Combustível: {self.tipo_combustivel}\n'
            f'Valor por Litro: R${self.valor_litro:.2f}\n'
            f'Quantidade Disponível: {self.quantidade_combustivel:.2f}')

def menu():
  bomba = BombaCombustivel(tipo_combustivel='Gasolina', valor_litro=5.59, quantidade_combustivel=1000)

  while True:
    print('\nMenu da Bomba de Combustível:')
    print('1. Abastecer por Valor')
    print('2. Abastecer por Litro')
    print('3. Alterar Valor do Combustível')
    print('4. Alterar Tipo do Combustível')
    print('5. Alterar Quantidade de Combustível na Bomba')
    print('6. Mostrar informações da Bomba')
    print('7. Sair')
    escolha = input('Escolha uma opção: ')

    if escolha == '1':
      valor = float(input('Informe o valor em reais para abastecer: '))
      bomba.abastecer_por_valor(valor)

    elif escolha == '2':
      litros = float(input('Informe a quantidade de litros para abastecer: '))
      bomba.abastecer_por_litro(litros)

    elif escolha == '3':
      novo_valor = float(input('Informe o novo valor por litro R$: '))
      bomba.alterar_valor(novo_valor)

    elif escolha == '4':
      novo_combustivel = input('Informe o novo tipo de combustível: ')
      bomba.alterar_combustivel(novo_combustivel)

    elif escolha == '5':
      nova_quantidade = input('Informe a nova quantidade de combustível na bomba (em litros): ')
      bomba.alterar_quantidade_combustivel(nova_quantidade)

    elif escolha == '6':
      print('\nInformações da Bomba:')
      print(bomba)

    elif escolha == '7':
      break

    else:
      print('Opção inválida! Tente Novamente!')
      continue

if __name__ == '__main__':
  menu()
````

````python
#Exercicio 11
class Carro:
  def __init__(self, consumo):
    self.consumo = consumo
    self.tanque = 0

  def adicionarGasolina(self, quantidade):
    self.tanque += quantidade
    print(f'{quantidade:.2f} litros de combustível adicionados ao tanque.')

  def andar(self, distancia):
    combustivel_necessario = distancia / self.consumo
    if combustivel_necessario <= self.tanque:
      self.tanque -= combustivel_necessario
      print(f'Você percorreu {distancia:.2f} km.')
    else:
      distancia_possivel = self.tanque * self.consumo
      self.tanque = 0
      print(f'Combustível insuficiente para percorrer {distancia:.2f} km.')
      print(f'Você conseguiu percorrer apenas {distancia_possivel:.2f} km antes de ficar sem combustível')

  def obterGasolina(self):
    return self.tanque

if __name__ == '__main__':
  meuFusca = Carro(15)
  meuFusca.adicionarGasolina(20)
  meuFusca.andar(100)
  gasolina_restante = meuFusca.obterGasolina()
  print(f'Combustível restante no tanque: {gasolina_restante:.2f} litros.')
````

````python
#Exercicio 12
class ContaInvestimento:
  def __init__(self, saldo_inicial=1000, taxa_juros=10):
    self.saldo = saldo_inicial
    self.taxa_juros = taxa_juros / 100

  def adicioneJuros(self):
    self.saldo += self.saldo * self.taxa_juros

  def __str__(self):
    return f'Saldo: R${self.saldo:.2f}, Taxa de Juros: {self.taxa_juros * 100:.2f}%'

if __name__ == '__main__':
  poupanca = ContaInvestimento(1000, 10)
  print('\nDados da Conta Poupança:')
  print(poupanca)

  for juro in range(5):
    poupanca.adicioneJuros()
  print('\nDados da Conta Poupança após 5 aplicações de juros:')
  print(poupanca)
````

````python
#Exercicio 17
import random

class BicinhoVirtual:
  def __init__(self, nome, fome=None, tedio=None):
    self.nome = nome
    self.fome = fome if fome is not None else random.randint(10, 90)
    self.tedio = tedio if tedio is not None else random.randint(10, 90)

  def alterarHumor(self):
    humor = self.fome + self.tedio
    if humor < 50:
      return 'Feliz 😊'
    elif humor < 100:
      return 'Ok 😐'
    else:
      return 'Triste 😢'

  def alimentar(self, quantidade):
    self.fome  -= quantidade
    if self.fome < 0:
      self.fome = 0

  def brincar(self, tempo):
    self.tedio -= tempo
    if self.tedio < 0:
      self.tideo = 0

  def passarTempo(self):
    self.fome += 5
    self.tedio += 5
    if self.fome > 100:
      self.fome = 100
    if self.tedio > 100:
      self. tedio = 100

  def __str__(self):
    return (f'{self.nome} | Fome {self.fome}/100 | Tédio: {self.tedio}/100 | Humor: {self.alterarHumor()}')


class FazendaBichinhos:
  def __init__(self, quantidade):
    self.bichinhos = [
        BicinhoVirtual(nome=f'Bicinho {i+1}') for i in range(quantidade)
    ]

  def alimentarTodos(self, quantidade):
    for bichinho in self.bichinhos:
      bichinho.alimentar(quantidade)
    print(f'Todos os bichinhos foram alimentados com {quantidade} unidades de comida!')

  def brincarTodos(self, tempo):
    for bichinho in self.bichinhos:
      bichinho.brincar(tempo)
    print(f'Todos os bichinhos brincaram por {tempo} minutos!')

  def passarTempo(self):
    for bichinho in self.bichinhos:
      bichinho.passarTempo()
    print('O tempo passou, e os bichinhos ficaram mais famintos e entendiados!')

  def mostrarStatusTodos(self):
    print('\nStatus dos Bichinhos na Fazenda:')
    for bichinho in self.bichinhos:
      print(bichinho)


def menu():
  print('Bem-vindo à Fazenda dos Bichinhos!')
  quantidade = int(input('Quantos bichinhos você quer criar na fazenda? '))
  fazenda = FazendaBichinhos(quantidade)

  while True:
    print('\nMenu do Tamagoshi:')
    print('1. Alimentar todos os bichinhos')
    print('2. Brincar com todos os bichinhos')
    print('3. Mostrar status de todos os bichinhos')
    print('4. Passar o tempo')
    print('5. Sair')
    escolha = input('Escolha uma opção: ')

    if escolha == '1':
      quantidade = int(input('Quanto de comida você quer fornecer para cada bichinho? '))
      fazenda.alimentarTodos(quantidade)

    elif escolha == '2':
      tempo = int(input('Por quanto tempo você quer brincar com cada bichinho (em minutos)? '))
      fazenda.brincarTodos(tempo)

    elif escolha == '3':
      fazenda.mostrarStatusTodos()

    elif escolha == '4':
      fazenda.passarTempo()

    elif escolha == '5':
      print('Encerrando o programa. Até a próxima!')
      break

    else:
      print('Opção inválida. Tente novamente!')

if __name__ == '__main__':
  menu()
````
