````python
#Exercicio 3
nome = input('Digite seu nome: ').strip()
while len(nome) < 3:
  print('Nome inválido! O nome deve conter pelo menos 3 caracteres.')
  nome = input('Digite seu nome: ').strip()

idade = int(input('Digite sua idade: '))
while idade < 0 or idade > 150:
  print('Idade inválida! A idade deve estar entre 0 e 150 anos.')
  idade = int(input('Digite sua idade: '))

salario = float(input('Digite seu salário: '))
while salario <= 0:
  print('Salário inválido! O salário deve ser maior que zero.')
  salario = float(input('Digite seu salário: '))

sexo = input('Digite seu sexo [F/M]: ').strip().lower()
while sexo not in ['f', 'm']:
  print('Sexo inválido! Digite "F" para Feminino ou "M" para Masculino.')
  sexo = input('Digite seu sexo [F/M]: ').strip().lower()

estado_civil = input('Digite seu estado civil [S, C, V, D]: ').strip().lower()
while estado_civil not in ['s', 'c', 'v', 'd']:
  print('Estado civil inválido! Digite "S" para Solteiro(a), "C" para Casado(a), "V" para Viúvo(a) ou "D" para Divorciado(a).')
  estado_civil = input('Digite seu estado civil [S, C, V, D]: ').strip().lower()

print(f"""
Nome: {nome}
Idade: {idade}
Salário R${salario:.2f}
Sexo: {"Feminino" if sexo == 'f' else "Masculino"}
Estado Civil: {"Solteiro(a)" if estado_civil == 's' else "Casado(a)" if estado_civil == 'c' else "Viúvo(a)" if estado_civil == 'v' else "Divorciado(a)"}""")
````

````python
#Exercicio 4
pais_A = 80000
pais_B = 200000
taxa_A = 0.03
taxa_B = 0.015
anos = 0

while pais_A <= pais_B:
  pais_A += pais_A * taxa_A
  pais_B += pais_B * taxa_B
  anos += 1

print(f"Serão necessários {anos} anos para que a população do país A ultrapasse ou iguale a população do país B.")
print(f"População do país A: {int(pais_A)}")
print(f"População do país B: {int(pais_B)}")
````

````python
#Exercicio 5
def calcular_populacao():
  while True:
    try:
      pais_A = float(input('Digite a população do país A: '))
      if pais_A <= 0:
        print('A população do país A deve ser um número positivo. Tente novamente.')
        continue
      pais_B = float(input('Digite a população do país B: '))
      if pais_B <= 0:
        print('A população do país B deve ser um número positivo. Tente novamente.')
        continue
      taxa_A = float(input('Digite a taxa de crescimento do país A (em %): ')) / 100
      if taxa_A <= 0:
        print('A taxa de crescimento do país A deve ser um número positivo. Tente novamente.')
        continue
      taxa_B = float(input('Digite a taxa de crescimento do país B (em %): ')) / 100
      if taxa_B <= 0:
        print('A taxa de crescimento do país B deve ser um número positivo. Tente novamente.')
        continue
      break
    except ValueError:
      print('Entrada inválida. Por favor, insira números válidos.')

  anos = 0

  while pais_A <= pais_B:
    pais_A += pais_A * taxa_A
    pais_B += pais_B * taxa_B
    anos += 1

  print(f"Serão necessários {anos} anos para que a população do país A ultrapasse ou iguale a população do país B.")
  print(f"População do país A: {int(pais_A)}")
  print(f"População do país B: {int(pais_B)}")

def repetir():
  while True:
    continuar = input('Deseja realizar outra operação? [S/N]: ').strip().lower()
    if continuar == 's':
      calcular_populacao()
    elif continuar == 'n':
      print('Programa encerrado.')
      break
    else:
      print('Opção inválida. Por favor, insira "S" para sim ou "N" para não.')

calcular_populacao()
repetir()
````

````python
#Exercicio 12
def tabuada(numero):
  print(f'Tabuada do {numero}:')
  for num in range(1, 11):
    print(f'{numero} x {num} = {num * numero}')

numero = int(input('Digite um número para a tabuada: '))
tabuada(numero)
````
