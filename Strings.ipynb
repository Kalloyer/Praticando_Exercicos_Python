## 14. Leet Speak Generator

**Leet** é uma forma de escrever o alfabeto latino usando outros símbolos em lugar das letras, como números. A própria palavra "leet" admite muitas variações, como **l33t** ou **1337**.

**Dica:** Pesquise sobre as principais formas de traduzir as letras para o formato leet.



````python
#Exercicio 1
def comparador_tamanho_str(string1, string2):
  try:
    print(f'\nString 1: {string1}')
    print(f'String 2: {string2}')
    print(f'Tamanho de: {string1}: {len(string1)} caracteres')
    print(f'Tamanho de: {string2}: {len(string2)} caracteres')

    if len(string1) != len(string2):
      print('As duas strings são de tamanhos diferentes.')
    else:
      print('As duas strings são de tamanhos iguais.')

    if string1 != string2:
      print('As duas strings possuem conteúdos diferentes.')
    else:
      print('As duas strings possuem conteúdos iguais.')
  except Exception as e:
        print(f"Error: {e}")

string1 = input('String 1: ')
string2 = input('String 2: ')
comparador_tamanho_str(string1, string2)
````

````python
#Exercicio 2
def nome_ao_contrario(nome):
  try:
    print(f'Nome: {nome}')
    print(f'Nome ao contrário: {nome[::-1]}')
  except Exception as e:
    print(f'Error: {e}')

nome = input('Digite um nome: ').strip().upper()
nome_ao_contrario(nome)
````

````python
#Exercicio 3
def nome_vertical(nome):
  print(f'Nome: {nome}')
  print(f'Nome na vertical:')
  for n in nome:
    try:
      print(f'{n}')
    except Exception as e:
      print(f'Error: {e}')

nome = input('Digite um nome: ').strip().upper()
nome_vertical(nome)
````

````python
#Exercicio 4
def nome_escada(nome):
  try:
    for n in range(1, len(nome) + 1):
      print(nome[:n])
  except Exception as e:
    print(f'Error: {e}')

nome = input('Digite um nome: ').strip().upper()
nome_escada(nome)
````

````python
#Exercicio 5
def nome_escada_invertido(nome):
  try:
    for n in range(len(nome), 0, - 1):
      print(nome[:n])
  except Exception as e:
    print(f'Error: {e}')

nome = input('Digite um nome: ').strip().upper()
nome_escada_invertido(nome)
````

````python
#Exercicio 6
from datetime import datetime

def data_por_extenso(data_str):
  meses = [
        "janeiro", "fevereiro", "março", "abril", "maio", "junho",
        "julho", "agosto", "setembro", "outubro", "novembro", "dezembro"
    ]
  try:
    data = datetime.strptime(data_str, "%d/%m/%Y")
    dia = data.day
    mes = meses[data.month - 1]
    ano = data.year
    return f'Você nasceu em {dia} de {mes.title()} de {ano}.'
  except ValueError:
    print('Data inválida. Digite uma data válida!')

data = input('Data de Nascimento [dd/mm/aaaa]: ')
resultado = data_por_extenso(data)
print(resultado)
````

````python
#Exercicio 7
def contador_espacos_vogais(palavra):
  try:
    contador_espacos = palavra.count(' ')
    contador_vogais = sum(1 for letra in palavra if letra in 'aeiou')
    print(f'\nQuantidade de espaços em branco na frase: {contador_espacos}')
    print(f'Quantidade de vogais na frase: {contador_vogais}')
  except Exception as e:
    print(f'Error: {e}')

palavra = input('Digite um palavra: ').lower()
contador_espacos_vogais(palavra)
````

````python
#Exercicio 8
import re

def palindromo(frase):
  try:
    frase_limpa = re.sub(r'[^a-zA-Z0-9]', '', frase).lower()
    if frase_limpa == frase_limpa[::-1]:
      print(f'\nA frase {frase} é um palíndromo.')
    else:
      print(f'\nA frase {frase} não é um palíndromo.')
  except Exception as e:
    print(f'Error: {e}')

frase = input('Digite uma frase: ').strip()
palindromo(frase)
````

````python
#Exercicio 9
import re

def validacao_CPF(documento):
  try:
    cpf_validado = re.sub(r'[^0-9]', '', documento)
    if len(cpf_validado) != 11:
      print(f'O CPF {documento} é inválido (deve conter 11 dígitos).')
      return
    if cpf_validado == cpf_validado[0] * 11:
      print(f'O CPF {documento} é inválido (todos os dígitos são iguais).')
      return

    soma1 = sum(int(cpf_validado[i]) * (10 - i) for i in range(9))
    digito1 = (soma1 * 10 % 11) % 10

    soma2 = sum(int(cpf_validado[i]) * (11 - i) for i in range(10))
    digito2 = (soma2 * 10 % 11) % 10

    if cpf_validado[-2:] == f'{digito1}{digito2}':
      print(f'O CPF {documento} é válido.')
    else:
      print(f'O CPF {documento} é inválido (dígitos verificadores incorretos).')
  except Exception as e:
    print(f'Erro ao processar o CPF: {e}')

documento = input('Digite seu CPF no formato xxx.xxx.xxx-xx: ').strip()
validacao_CPF(documento)
````

````python
#Exercicio 10
def numero_extenso(numero):
  unidades = [
        "zero", "um", "dois", "três", "quatro",
        "cinco", "seis", "sete", "oito", "nove"
    ]
  especiais = [
      "dez", "onze", "doze", "treze", "catorze",
      "quinze", "dezesseis", "dezessete", "dezoito", "dezenove"
    ]
  dezenas = [
      "", "", "vinte", "trinta", "quarenta",
      "cinquenta", "sessenta", "setenta", "oitenta", "noventa"
    ]

  if numero < 0 or numero > 99:
    return 'Número fora do intervalo. Digite um número entre 0 e 99.'
  if 10 <= numero <= 19:
    return especiais[numero - 10]
  if numero < 10:
    return unidades[numero]
  if numero % 10 == 0:
    return dezenas[numero // 10]

  dezena = dezenas[numero // 10]
  unidade = unidades[numero % 10]
  return f'{dezena} e {unidade}'

try:
  numero = int(input('Digite um número entre 0 e 99:'))
  print(f'O número {numero} por extenso é: {numero_extenso(numero)}')
except ValueError:
  print('Entrada inválida! Por favor, digite um número inteiro.')
````

````python
#Exercicio 11
import random
def carregar_palavras(arquivo):
  try:
    with open(arquivo, 'r') as f:
      palavras = f.read().strip().splitlines()
    return [palavra.strip().upper() for palavra in palavras if palavra.strip()]
  except FileNotFoundError:
    print('Arquivo de palavras não encontrado!')
    return []

def jogo_forca():
  palavras = carregar_palavras('palavras_forca.txt')
  if not palavras:
    print("Nenhuma palavra disponível para o jogo. Verifique o arquivo 'palavras.txt'.")
    return

  palavra_secreta = random.choice(palavras)
  palavra_oculta = ['_'] * len(palavra_secreta)
  tentativas_erradas = 0
  letras_erradas = []
  max_tentativas = 6

  print('\nBem-vindo ao Jogo da Forca!')
  print(f'A palavra tem {len(palavra_secreta)} letras.')
  print(' '.join(palavra_oculta))

  while tentativas_erradas < max_tentativas:
    letra = input('\nDigite uma letra: ').strip().upper()
    if len(letra) != 1 or not letra.isalpha():
      print('Por favor, digite apenas uma letra.')
      continue
    if letra in letras_erradas or letra in palavra_oculta:
      print('Você já tentou essa letra. Tente outra.')
      continue
    if letra in palavra_secreta:
      print(f"Boa! A letra '{letra}' está na palavra.")
      for i, l in enumerate(palavra_secreta):
        if l == letra:
          palavra_oculta[i] = letra
    else:
      tentativas_erradas += 1
      letras_erradas.append(letra)
      print(f'-> Você errou pela {tentativas_erradas}ª vez. Tente de novo!')

    print("A palavra é: " + " ".join(palavra_oculta))
    if '_' not in palavra_oculta:
      print('\nParabéns! Você descobriu a palavra e venceu o jogo!')
      break
  if tentativas_erradas == max_tentativas:
    print("\nVocê foi enforcado! A palavra era:", palavra_secreta)

  print('Fim do jogo!')

if __name__ == '__main__':
  jogo_forca()
````

````python
#Exercicio 12
def corrige_telefone(numero):
  numero = numero.replace('-', '').replace(' ', '')
  if len(numero) == 7:
    print('\nTelefone possui 7 dígitos. Vou acrescentar o dígito três na frente.')
    numero_corrigido = '3' + numero
  else:
    numero_corrigido = numero

  if len(numero_corrigido) == 8:
    print(f'Telefone corrigido sem formatação: {numero_corrigido}')
    numero_formatado = f'{numero_corrigido[:4]}-{numero_corrigido[4:]}'
    print(f'Telefone corrigido com formatação: {numero_formatado}')
  else:
    print(f'Erro: O número {numero} não é válido. Deve ter 7 ou 8 dígitos após a correção.')

telefone = input('Digite um número de telefone: ').strip()
corrige_telefone(telefone)
````

````python
#Exercicio 13
import random

def carregar_arquivo(arquivo):
  try:
    with open(arquivo) as f:
      palavras = f.read().strip().splitlines()
    return [palavra.strip().upper() for palavra in palavras if palavra.strip()]
  except FileNotFoundError:
    print('Arquivo de palavras não encontrado!')
    return []

def palavra_embaralhada():
  palavras = carregar_arquivo('palavras_embaralhadas.txt')
  if not palavras:
    print("Nenhuma palavra disponível para o jogo. Verifique o arquivo palavras_embaralhadas.txt.")
    return
  palavra_secreta = random.choice(palavras)
  palavra_embaralhada = list(palavra_secreta)
  random.shuffle(palavra_embaralhada)
  palavra_embaralhada = ''.join(palavra_embaralhada)

  print('\nBem-vindo ao Jogo da Palavra Embaralhada!')
  print(f'A palavra embaralhada é: {palavra_embaralhada}')
  print('Você tem 6 tentativas para adivinhar a palavra.')

  tentativas = 6
  while tentativas > 0:
    resposta = input('Digite sua tentativa: ').strip().upper()
    if resposta == palavra_secreta:
      print('\nParabéns! Você acertou a palavra!')
      print(f'A palavra era: {palavra_secreta}')
      return
    else:
      tentativas -= 1
      print(f'Resposta incorreta! Você ainda tem {tentativas} tentativa(s).')
  print('\nVocê perdeu! Suas tentativas acabaram.')
  print('A palavra era: {palavra_secreta}')

if __name__ == '__main__':
  palavra_embaralhada()
````

````python
#Exercicio 14
def leet_speak(texto):
  leet_dict = {
    'A': '4', 'B': '8', 'C': '<', 'D': '|)', 'E': '3',
    'F': '|=', 'G': '6', 'H': '#', 'I': '1', 'J': '_|',
    'K': '|<', 'L': '1', 'M': '/\\/\\', 'N': '|\\|',
    'O': '0', 'P': '|*', 'Q': '0_', 'R': '|2', 'S': '5',
    'T': '7', 'U': '|_|', 'V': '\\/', 'W': '\\/\\/',
    'X': '><', 'Y': '`/', 'Z': '2'
  }
  texto_leet = ''.join(leet_dict.get(char.upper(), char) for char in texto)
  return texto_leet

texto_usuario = input('Digite um texto para converter para Leet Speak: ').strip()
resultado = leet_speak(texto_usuario)

print(f'Texto original: {texto_usuario}')
print(f'Texto em Leet Speak: {resultado}')
````
