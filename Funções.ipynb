```python
#Exercicio 5
def somaImposto(n):
  taxaImposto = 10 / 100
  venda = n
  valor_final = venda + (venda * taxaImposto)
  print(f'\nO valor de venda do produto é de R${venda:.2f}')
  print(f'O valor do produto com imposta é de R${valor_final:.2f}')

try:
  n = float(input('Digite o valor do produto R$: '))
  if n > 0:
    somaImposto(n)
  else:
    print('Por favor, insira um valor positivo.')
except ValueError:
  print('Entrada inválida! Digite um valor positivo.')
```

```python
#Exercicio 6
def converter_12_horas(horas, minutos):
  if horas == 0:
    horas_12 = 12
    periodo = 'A'
  elif horas == 12:
    horas_12 = 12
    periodo = 'P'
  elif horas > 12:
    horas_12 = horas - 12
    periodo = 'P'
  else:
    horas_12 = horas
    periodo = 'A'
  return horas_12, minutos, periodo

def exibir_12_horas(horas_12, minutos, periodo):
  periodo_extenso = 'A.M.' if periodo == 'A' else 'P.M.'
  print(f'{horas_12}:{minutos:02d} {periodo_extenso}')

def main():
  while True:
    try:
      horas = int(input('Digite a hora (0 a 23): '))
      minutos = int(input('Digite os minutos (0 a 59): '))
      if horas < 0 or horas > 23 or minutos < 0 or minutos > 59:
        print('Por favor, insira valores válidos para horas (0-23) e minutos (0-59).')
        continue

      horas_12, minutos_12, periodo = converter_12_horas(horas, minutos)

      print('\nHora do formato 12 horas: ')
      exibir_12_horas(horas_12, minutos_12, periodo)

      continuar = input('Deseja converter outra hora? (s/n): ').strip().lower()
      if continuar != 's':
        print('Programa encerrado.')
        break
    except ValueError:
      print('Entrada inválida! Digite apenas números inteiros para as horas e minutos.')

main()
```

```python
#Exercicio 7
def valorPagamento(prestacao, dias_atraso):
  multa = 3 / 100
  juros = 0.1 / 100

  if dias_atraso > 0:
    valor = prestacao + (prestacao * multa) + (prestacao * juros * dias_atraso)
  else:
    valor = prestacao
  return valor

def main():
  total_pago = 0
  quantidade_prestacoes = 0

  while True:
    try:
      prestacao = float(input('Digite o valor da prestação R$ (0=fim): '))
      if prestacao == 0:
        break
      dias_atraso = int(input('Digite o número de dias de atraso (0 se não houver): '))
      if dias_atraso < 0:
        print('Os dias de atraso não podem ser negativos.')
        continue

      valor_final = valorPagamento(prestacao, dias_atraso)
      total_pago += valor_final
      quantidade_prestacoes += 1
    except ValueError:
      print('Entrada inválida! Digite números válidos para a prestação e dias de atraso.\n')

  print(f"""
Relatório do dia:
Quantidade de prestações pagas: {quantidade_prestacoes}
Valor total pago: R${total_pago:.2f}""")

main()
```

```python
#Exercicio 9
def numero_reverso():
  while True:
    try:
      numero = int(input('Digite um número: '))
      if numero < 0:
        print('Por favor, insira um número inteiro.')
        continue
      break
    except ValueError:
      print('Entrada inválida! Por favor, digite um número inteiro.')

  numero_invertido = int(str(numero)[::-1])
  print(f'n\Número digitado: {numero}')
  print(f'Número digitado invertido: {numero_invertido}')

numero_reverso()
```

```python
#Exercicio 10
import random
import time

def lancar_dados():
  dado1 = random.randint(1, 6)
  dado2 = random.randint(1, 6)
  return dado1 + dado2

def jogo_de_craps():
  print('Bem-vindo ao jogo de Craps!')
  while True:
    print('\nLançando os dados...')
    time.sleep(2.5)
    resultado = lancar_dados()
    print(f'Você tirou: {resultado}')

    if resultado in (7, 11):
      print("Parabéns! Você tirou um 'natural' e ganhou!")
    elif resultado in (2, 3, 12):
      print("Que pena! Você tirou 'craps' e perdeu!")
    else:
      ponto = resultado
      print(f'Seu Ponto é: {ponto}. Agora você deve tirar {ponto} novamente antes de tirar 7.')
      while True:
        print('\nLançando os dados...')
        time.sleep(2.5)
        resultado = lancar_dados()
        print(f'Você tirou: {resultado}')
        if resultado == ponto:
          print('Parabéns! Você tirou seu Ponto ({ponto}) e ganhou!')
          break
        elif resultado == 7:
          print('Que pena! Você tirou 7 e perdeu!')
          break

    continuar = input('\nDeseja jogar novamente? (s/n): ').strip().lower()
    if continuar != 's':
      print('Obrigado por jogar! Até a próxima.')
      break

jogo_de_craps()
```

```python
#Exercicio 11
from datetime import datetime

def data_por_extenso(data_str):
  meses = [
        "janeiro", "fevereiro", "março", "abril", "maio", "junho",
        "julho", "agosto", "setembro", "outubro", "novembro", "dezembro"
    ]
  try:
    data = datetime.strptime(data_str, '%d/%m/%Y')
    dia = data.day
    mes = meses[data.month - 1]
    ano = data.year
    return f'{dia} de {mes} de {ano}'
  except ValueError:
    print('NULL')

data_entrada = input('Difite a data no formato DD/MM/AAAA: ')
resultado = data_por_extenso(data_entrada)
print(resultado)
```

```python
#Exercicio 12
from random import shuffle

def embaralha_palavra(palavra):
  try:
    palavra = palavra.strip().lower()
    lista_caracteres = list(palavra)
    shuffle(lista_caracteres)
    palavra_embaralhada = ''.join(lista_caracteres)
    print(f'\nPalavra original: {palavra}')
    print(f'Palavra embaralhada: {palavra_embaralhada}')
  except Exception as e:
    print(f'Erro: {e}.')

palavra = input('Digite uma palavra: ')
embaralha_palavra(palavra)
```

```python
#Exercicio 13
def desenho_moldura(linhas=1, colunas=1):
  linhas = max(1, min(linhas, 20))
  colunas = max(1, min(colunas, 20))
  try:
    print('+' + '-' * (colunas - 2) + '+')
    for _ in range(linhas - 2):
      print('|' + ' ' * (colunas - 2) + '|')

    if linhas > 1:
      print( '+' + '-' * (colunas - 2) + '+')
  except Exception as e:
    print(f'Erro ao desenhar a moldura: {e}.')

try:
  linhas = int(input('Digite a quantidade de linhas: '))
  colunas = int(input('Digite a quantidade de colunas: '))
  desenho_moldura(linhas, colunas)
except ValueError:
  print('Entrada inválida! Digite números inteiros entre 1 e 20.')
```

```python
#Exercicio 14
from itertools import permutations

def verifica_quadrado_magico(quadrado):
  soma_magica = 15

  for i in range(0, 9, 3):
    if sum(quadrado[i:i+3]) != soma_magica:
      return False

  for i in range(3):
    if quadrado[i] + quadrado[i+3] + quadrado[i+6] != soma_magica:
      return False

  if quadrado[0] + quadrado[4] + quadrado[8] != soma_magica:
      return False
  if quadrado[2] + quadrado[4] + quadrado[6] != soma_magica:
      return False
  return True

def encontra_quadrados_magicos():
  numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9]
  permutacoes = permutations(numeros)

  quadrados_magicos = []

  for p in permutacoes:
    if verifica_quadrado_magico(p):
      quadrados_magicos.append(p)

  print('Encontrados {len(quadrados_magicos)} quadrados mágicos 3x3:\n')
  for quadrado in quadrados_magicos:
    print(f'{quadrado[0]}   {quadrado[1]}   {quadrado[2]}')
    print(f'{quadrado[3]}   {quadrado[4]}   {quadrado[5]}')
    print(f'{quadrado[6]}   {quadrado[7]}   {quadrado[8]}')
    print('-' * 10)

encontra_quadrados_magicos()
```
