````python
#Exercicio 12
def folha_de_pagamento(salario):
  if salario <= 900:
    desconto_ir = 0
  elif salario <= 1500:
    desconto_ir = 5
  elif salario <= 2500:
    desconto_ir = 10
  else:
    desconto_ir = 20

  ir = salario * (desconto_ir / 100)
  inss = salario * 0.1
  fgts = salario * 0.11
  sindicato = salario * 0.03
  total_desconto = ir + inss + sindicato
  salario_liquido = salario - total_desconto

  return f"""Salário Bruto: R${salario:.2f}
  IR ({desconto_ir}%): R${ir:.2f}
  INSS (10%): R${inss:.2f}
  Sindicato (3%): R${sindicato:.2f}
  FGTS (11%): R${fgts:.2f}
  Total de descontos: R${total_desconto:.2f}
  Salário Líquido: R${salario_liquido:.2f}"""

print(folha_de_pagamento(900))
print(folha_de_pagamento(2000))
print(folha_de_pagamento(3000))
````

````python
# Exercicio 15
def triangulo(lado1, lado2, lado3):
  if (lado1 + lado2 > lado3) and (lado2 + lado3 > lado1) and (lado1 +lado3 > lado2):
    print('É um triângulo!')
    if lado1 == lado2 == lado3:
      print('Triângulo Equilátero')
    elif lado1 == lado2 or lado2 == lado3 or lado1 == lado3:
      print('Triângulo Isósceles')
    else:
      print('Triângulo Escaleno')
  else:
    print('Não é um triângulo')

triangulo(1, 3, 2)  # Não é um triângulo
triangulo(12, 3, 6)  # Não é um triângulo
triangulo(3, 3, 3)  # É um triângulo, Triângulo Equilátero
triangulo(5, 5, 8)  # É um triângulo, Triângulo Isósceles
triangulo(3, 4, 5)  # É um triângulo, Triângulo Escaleno
````

````python
#Exercicio 19
def decompor_numero(numero):
  if numero < 0 or numero >= 1000:
    print('Por favor, insira um número inteiro entre 0 e 999.')
    return

  centena = numero // 100
  resto_numero = numero % 100
  dezena = resto_numero // 10
  unidade = resto_numero % 10

  print(f'O número {numero} é composto por: ')
  if centena > 0:
    print(f'{centena} centena(s)')
  if dezena > 0:
    print(f'{dezena} dezena(s)')
  if unidade > 0:
    print(f'{unidade} unidade(s)')

numero = int(input('Digite um número inteiro menor que 1000: '))
decompor_numero(numero)
````

````python
#Exercicio 24
def calculo(opcao, numero1, numero2):
  if opcao == '+':
    return numero1 + numero2
  elif opcao == '-':
    return numero1 - numero2
  elif opcao == '/':
    return numero1 / numero2
  elif opcao == '*':
    return numero1 * numero2
  else:
    return 'Operação Inválida.'

def operacao(numero):
  if numero % 2 == 0:
    print(f'O número {numero:.2f} é par.')
  else:
    print(f'O número {numero:.2f} é impar.')

  if numero >= 0:
    print(f'O número {numero:.2f} é positivo.')
  else:
    print(f'O número {numero:.2f} é negativo.')

  if numero == round(numero):
      print(f'O número {numero:.2f} é inteiro.')
  else:
    print(f'O número {numero:.2f} é decimal.')

numero1 = float(input('Digite o primeiro numero: '))
numero2 = float(input('Digite o segundo numero: '))
opcao = str(input('Digite um opção [+, -, /, *], para o cálculo: '))

resultado = calculo(opcao, numero1, numero2)
if isinstance(resultado, str):
  print(resultado)
else:
  print(f'O resultado da operação {numero1} {opcao} {numero2} é: {resultado:.2f}')

  operacao(resultado)
````

````python
#Exercicio 25
def crime(respostas):
  repostas_positivas = sum(respostas)

  if repostas_positivas == 2:
    return 'Suspeita'
  elif 3 <= repostas_positivas <= 4:
    return 'Cúmplice'
  elif repostas_positivas == 5:
    return 'Assassino'
  else:
    return 'Inocente'

perguntas = [
    'Telefonou para a vítima? [S/N]: ',
    'Esteve no local do crime? [S/N]: ',
    'Mora perto da vítima? [S/N]: ',
    'Devia para a vítima? [S/N]: ',
    'Já trabalhou com a vítima? [S/N]: '
]

respostas = []
for pergunta in perguntas:
  while True:
    resposta = input(pergunta).strip().upper()
    if resposta in ['S', 'N']:
      respostas.append(1 if resposta == 'S' else 0)
      break
    else:
      print("Resposta inválida! Por favor, digite 'S' para Sim ou 'N' para Não.")

resultado = crime(respostas)
print(f'\nClassificação: {resultado}')
````

````python
#Exercicio 26
def calculo_combustivel(tipo, litros):
  if tipo == 'A':
    preco_alcool = 1.9
    if litros <= 20:
      desconto = 0.03
    else:
      desconto = 0.05
    valor = litros * (preco_alcool - (preco_alcool * desconto))
  elif tipo == 'G':
    preco_gasolina = 2.5
    if litros <= 20:
      desconto = 0.04
    else:
      desconto = 0.06
    valor = litros * (preco_gasolina - (preco_gasolina * desconto))
  else:
    return 'Tipo de combustível inválido!'

  return f'Valor do abastecimento: R${valor:.2f}'

tipo = input('Digite o tipo de combústivel [A-álcool, G-gasolina]: ').strip().upper()
litros = float(input('Digite a quantidade de litros abastecidos: '))

resultado = calculo_combustivel(tipo, litros)
print(resultado)
````

````python
#Exercicio 27
def calculo_kg(fruta, kilo):
  if fruta == 'MACA':
    preco = 2.5 if kilo <= 5 else 2.2
  elif fruta == 'MORANGO':
    preco = 1.8 if kilo <= 5 else 1.5
  else:
    return 'Tipo de fruta inválida!'

  valor = preco * kilo

  if kilo > 8 or valor > 25:
      desconto = 0.1
      valor -= valor * desconto

  return f'Preço total para {kilo:.2f} kg de {fruta}: R${valor:.2f}'

fruta = input('Digite qual fruta irá comprar [Maca ou Morango]: ').strip().upper()
kilo = float(input('Digite a quantidade de kilos da fruta: '))

resultado = calculo_kg(fruta, kilo)
print(resultado)
````

````python
#Exercicio 28
def hipermercado_tabajara(carne, kilo, pagamento):
  if carne == 'FILE DUPLO':
    preco = 4.9 if kilo <= 5 else 5.8
  elif carne == 'ALCATRA':
    preco = 5.9 if kilo <= 5 else 6.8
  elif carne == 'PICANHA':
    preco = 6.9 if kilo <= 5 else 7.8
  else:
    return 'Opção Inváida!'

  valor_carne = preco * kilo

  desconto = 0
  if pagamento.upper() == 'CARTAO TABAJARA':
    desconto = valor_carne * 0.1

  valor_total = valor_carne - desconto

  return f"""CUPOM FISCAL
Carne: {carne}
Kilos: {kilo:.2f} kg
Preço Total: R${valor_carne:.2f}
Tipo de Pagamento: {pagamento}
Desconto: R${desconto:.2f}
Valor Total: R${valor_total:.2f}"""

carne = input('Digite qual carne foi escolida [File Duplo, Alcatra, Picanha]: ').strip().upper()
kilo = float(input('Digite a quantidade de quilos de carne: '))
pagamento = input('Digite o tipo de pagamento [Cartao Tabajara, Dinheiro, Pix]: ').strip()

resultado = hipermercado_tabajara(carne, kilo, pagamento)
print(resultado)
````
