24.  Faça um programa que simule um lançamento de dados. Lance o dado 100 vezes e armazene os resultados em um vetor . Depois, mostre quantas vezes cada valor foi conseguido. Dica: use um vetor de contadores(1-6) e uma função para gerar numeros aleatórios, simulando os lançamentos dos dados.


```python
#Exercicio 1
def ler_vetor5():
  vetor = []

  print('Digite 5 números inteiros:')
  for n in range(1, 6):
    while True:
      try:
        numero = int(input(f'Digite o {n}º número: '))
        vetor.append(numero)
        break
      except ValueError:
        print('ntrada inválida! Por favor, digite um número inteiro.')

  print('\nOs números digitados foram:', end=' ')
  for numero in vetor:
    print(numero, end= ' ')

ler_vetor5()
```

```python
#Exercicio 2
def ler_vetor10():
  vetor = []

  print('Digite 10 números inteiros:')
  for n in range(1, 11):
    while True:
      try:
        numero = int(input(f'Digite o {n}º número: '))
        vetor.append(numero)
        break
      except ValueError:
        print('ntrada inválida! Por favor, digite um número inteiro.')

  print('\nOs números digitados foram:', end=' ')
  for numero in vetor[::-1]:
    print(numero, end= ' ')

ler_vetor10()
```

```python
#Exercicio 3
def leia_4notas():
  notas = []

  for i in range(1, 5):
    while True:
      try:
        nota = float(input(f'Digite a {i}º nota: '))
        if 0 <= nota <= 10:
          notas.append(nota)
          break
        else:
          print('Por favor, digite uma nota válida entre 0 e 10.')
      except ValueError:
        print('Entrada inválida! Por favor, digite um número inteiro.')

  media_notas = sum(notas) / len(notas)

  print('\nNotas digitadas: ')
  for i, nota in enumerate(notas, start=1):
    print(f'{i}ª Nota: {nota:.2f}')
  print(f'Média das notas: {media_notas:.2f}')

leia_4notas()
```

```python
#Exercicio 4
def vetor_10caracteres():
  consoantes = []
  vogais = 'aeiou'

  for i in range(1, 11):
    while True:
      try:
        letra = input(f'Digite o {i}º caractere: ').strip().lower()
        if len(letra) == 1 and letra.isalpha():
          if letra not in vogais:
            consoantes.append(letra)
          break
      except ValueError:
        print('Entrada inválida! Por favor, digite um vetor.')

  print(f'\nForam lidas {len(consoantes)} consoantes.')
  print('Consoantes digitadas:', ','.join(consoantes))

vetor_10caracteres()
```

```python
#Exercicio 5
def armazem_vetor():
  vetores = []
  vetores_par = []
  vetores_impar = []

  for i in range(1, 21):
    while True:
      try:
        numero = int(input('Digite um número inteiro: '))
        if numero >= 0:
          vetores.append(numero)
          if numero % 2 == 0:
            vetores_par.append(numero)
          elif numero % 1 == 0:
            vetores_impar.append(numero)
          break
        else:
          print('Por favor, digite um número maior que 0.')
      except ValueError:
        print('Entrada inválida! Por favor, digite um vetor.')

  print(f"""
Números digitados: {vetores}
Números pares digitados: {vetores_par}
Números ímpares digitados: {vetores_impar}""")

armazem_vetor()
```

```python
#Exercicio 6
def nota_10alunos():
  medias = []
  num_alunos = 10

  for aluno in range(1, num_alunos + 1):
    notas = []
    print(f'\nAluno(a): {aluno}')
    for i in range(1, 5):
      while True:
        try:
          nota = float(input(f'Digite a {i}ª nota: '))
          if 0 <= nota <= 10:
            notas.append(nota)
            break
          else:
            print('A nota deve estar entre 0 e 10.')
        except ValueError:
          print('Entrada inválida! Por favor, digite um número.')

    media = sum(notas) / 4
    medias.append(media)

  aprovados = [media for media in medias if media > 7]
  print('\nMédia dos alunos: ')
  for i, media in enumerate(medias, start=1):
    print(f'Aluno(a) {i}: {media:.2f}')
  print(f'\nNúmero de alunos aprovados com média maior ou igual a 7: {len(aprovados)}')

nota_10alunos()
```

```python
#Exercicio 7
def somar_multiplicar_5vetor():
  vetor = []
  multiplicao_vetor = 1
  soma_vetor = 0

  print('Digite 5 números inteiros:')
  for n in range(1, 6):
    while True:
      try:
        numero = int(input(f'Digite o {n}º número: '))
        vetor.append(numero)
        soma_vetor += numero
        multiplicao_vetor *= numero
        break
      except ValueError:
        print('Entrada inválida! Por favor, digite um número inteiro.')

  print('\nOs números digitados foram:', end=' ')
  for numero in vetor:
    print(numero, end= ' ')
  print(f'\nA soma dos número é de {soma_vetor}')
  print(f'A multiplicação dos número é de {multiplicao_vetor}')
somar_multiplicar_5vetor()
```

```python
#Exercicio 8
def idade_altura5():
  idades = []
  alturas = []

  for i in range(1, 6):
    while True:
      try:
        idade = int(input(f'Digite a idade da {i}ª pessoa: '))
        altura = float(input(f'Digite a altura em metros da {i}ª pessoa: '))
        if idade >= 0 and altura >= 0:
          idades.append(idade)
          alturas.append(altura)
          break
        else:
          print('Insira valores maiores ou iguais a 0.')
      except ValueError:
        print('Entrada inválida! Por favor, digite um número inteiro.')

  print("\nAs idades e alturas inseridas na ordem inversa foram:")
  print("Idades: ", idades[::-1])
  print("Alturas: ", alturas[::-1])

idade_altura5()
```

```python
#Exercicio 10
def intercalar_vetores():
  vetor1 = []
  vetor2 = []
  vetor3 = []

  print('Digite os 10 elementos do primeiro vetor: ')
  for i in range(1, 11):
    while True:
      try:
        numero = int(input(f'Digite o {i}º número do vetor 1: '))
        vetor1.append(numero)
        break
      except ValueError:
        print('Entrada inválida! Por favor, digite um número inteiro.')

  print('\nDigite os 10 elementos do segundo vetor: ')
  for i in range(1, 11):
    while True:
      try:
        numero = int(input(f'Digite o {i}º número do vetor 1: '))
        vetor2.append(numero)
        break
      except ValueError:
        print('Entrada inválida! Por favor, digite um número inteiro.')

  for i in range(10):
    vetor3.append(vetor1[i])
    vetor3.append(vetor2[i])
  print('\nO terceiro vetor intercalado é:', end=' ')
  print(vetor3)

intercalar_vetores()
```

```python
#Exercicio 13
def media_anual_temperatura():
  meses = [
      'Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
      'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'
  ]
  temperaturas = []

  for i, mes in enumerate(meses):
    while True:
      try:
        temperatura = float(input(f'Digite a temperatura de {mes}: '))
        temperaturas.append(temperatura)
        break
      except ValueError:
        print('Entrada inválida! Por favor, digite um número inteiro.')

  media_anual = sum(temperaturas) / len(temperaturas)

  print(f"\nA média anual das temperaturas é: {media_anual:.2f}°C")
  print("Meses com temperaturas acima da média anual:")
  for i, temperatura in enumerate(temperaturas):
    if temperatura > media_anual:
      print(f'{i + 1} - {meses[i]}: {temperatura:.2f}')

media_anual_temperatura()
```

```python
#Exercicio 17
def salto_distancia():
  while True:
    atleta = input("Digite o nome do(a) atleta (ou deixe vazio para encerrar): ").strip()
    if not atleta:
      print("Programa encerrado.")
      break

    saltos = []

    print(f"\nDigite as distâncias dos 5 saltos para o(a) atleta {atleta}:")
    for i in range(1, 6):
      while True:
        try:
          salto = float(input(f"Digite o {i}º salto (em metros): "))
          if salto <= 0:
            print("A distância deve ser maior que zero. Tente novamente.")
            continue
          saltos.append(salto)
          break
        except ValueError:
          print("Entrada inválida! Digite um número válido.")

    media_saltos = sum(saltos) / len(saltos)
    saltos_formatados = ' - '.join(f'{salto:.1f}' for salto in saltos)

    print(f"""
Atleta: {atleta}

Primeiro Salto: {saltos[0]:.1f} m
Segundo Salto: {saltos[1]:.1f} m
Terceiro Salto: {saltos[2]:.1f} m
Quarto Salto: {saltos[3]:.1f} m
Quinto Salto: {saltos[4]:.1f} m

Resultado final:
{atleta}: {media_saltos:.1f} m
Saltos: {saltos_formatados}
Média dos demais saltos: {media_saltos:.1f} m
""")

salto_distancia()
```

```python
#Exercicio 18
def calcular_percentual(votos_jogador, total_votos):
  return (votos_jogador / total_votos) * 100 if total_votos > 0 else 0


def enquete_emissora_jogadores():
  jogadores = [0] * 23
  total_votos = 0

  print('Enquete: Quem foi o melhor jogador?')
  while True:
    try:
      voto = int(input('Número do jogador (0=fim): '))
      if voto == 0:
        break
      if 1 <= voto <= 23:
        jogadores[voto - 1] += 1
        total_votos += 1
      else:
        print('Informe um valor entre 1 e 23 ou 0 para sair!')
    except ValueError:
      print('Entrada inválida! Digite um número entre 1 e 23.')
  if total_votos == 0:
    print('Nenhum voto computado.')
    return

  print('\nResultado da votação:')
  print(f'\nForam computados {total_votos} votos.\n')
  print(f"{'Jogador':<10} {'Votos':<10} {'%':<10}")

  melhor_jogador = -1
  max_votos = 0
  for i, votos in enumerate(jogadores):
    if votos > 0:
      percentual = calcular_percentual(votos, total_votos)
      print(f"{i + 1:<10}{votos:<8}{percentual:.1f}%")
      if votos > max_votos:
        max_votos = votos
        melhor_jogador = i + 1

  percentual_melhor = calcular_percentual(max_votos, total_votos)
  print(f'O melhor jogador foi o número {melhor_jogador}, com {max_votos} votos, correspondendo a {percentual_melhor:.1f}% do total de votos.')

  with open('resultado_votacao.txt', 'w') as arquivo:
    arquivo.write('\nResultado da votação:')
    arquivo.write(f'\nForam computados {total_votos} votos.')
    arquivo.write(f"{'Jogador':<10} {'Votos':<10} {'%':<10}")
    for i, votos in enumerate(jogadores):
      if votos > 0:
        percentual = calcular_percentual(votos, total_votos)
        arquivo.write(f"{i + 1:<10} {votos:<10} {percentual:.1f}%")
    arquivo.write(f'O melhor jogador foi o número {melhor_jogador}, com {max_votos} votos, correspondendo a {percentual_melhor:.1f}% do total de votos.')
  print("\nResultados gravados no arquivo 'resultado_votacao.txt'.")

enquete_emissora_jogadores()
```

```python
#Exercicio 19
def calcular_percentual(votos_usuario, total_votos):
  return (votos_usuario / total_votos) * 100 if total_votos > 0 else 0

def enquete_sistema_operacional():
  sistema_operacioanl = [0] * 6
  nomes_sistemas = [
    "Windows Server", "Unix", "Linux", "Netware", "Mac OS", "Outro"
  ]
  total_votos = 0

  print('Qual o melhor Sistema Operacional para uso em servidores?')
  print("""
As possíveis respostas são:

1- Windows Server
2- Unix
3- Linux
4- Netware
5- Mac OS
6- Outro""")

  while True:
    try:
      voto = int(input('Escolha sua opção (0=fim): '))
      if voto == 0:
        break
      if 1 <= voto <= 6:
        sistema_operacioanl[voto - 1] += 1
        total_votos += 1
      else:
        print('Informe um valor entre 1 e 6 ou 0 para sair!')
    except ValueError:
      print('Entrada inválida! Digite um número entre 1 e 6.')

  if total_votos == 0:
    print('Nenhum voto computado.')
    return

  print(f"\n{'Sistema Operacional':<20} {'Votos':<20} {'%':<20}")
  print(f"{'-------------------':<20} {'-----':<20} {'-':<20}")

  melhor_sistema = -1
  max_votos = 0

  for i, votos in enumerate(sistema_operacioanl):
    if votos > 0:
      percentual = calcular_percentual(votos, total_votos)
      print(f"{nomes_sistemas[i]:<20} {votos:<20} {percentual:.2f}%")
      if votos > max_votos:
        max_votos = votos
        melhor_sistema = nomes_sistemas[i]

  print(f"{'-------------------':<20} {'-----':<20}")
  print(f"{'Total':<20} {total_votos:<20}")

  percentual_melhor = calcular_percentual(max_votos, total_votos)
  print(f'\nO Sistema Operacional mais votado foi o {melhor_sistema}, com {max_votos} votos, correspondendo a {percentual_melhor:.1f}% dos votos.')

enquete_sistema_operacional()
```

```python
#Exercicio 20
def abono_colaboradores():
  salarios = []
  abonos = []

  print('Projeção de Gastos com Abono')
  print('============================\n')

  while True:
    try:
      salario = float(input('Salário: '))
      if salario <= 0:
        break
      salarios.append(salario)
      if salario < 1000:
        abonos.append(100)
      elif salario >= 1000:
        abonos.append(salario * 0.2)
      else:
        print('Entrada inválida! Digite um valor maior que 0.')
    except ValueError:
      print('Entrada inválida! Digite um valor maior que 0.')

  salarios_processados = len(salarios)
  total_abono = sum(abonos)
  abono_minimo = len([abono for abono in abonos if abono <= 100])
  maior_abono = max(abonos)

  print(f"\n{'Salario':<10}{'-':^10}{'Abono':<20}")
  for i, abono in enumerate(salarios):
    if abono > 0:
      print(f"R${salarios[i]:<10}{'-':^10}R${abonos[i]:<5.2f}")

  print(f"""
Foram processados {salarios_processados} colaboradores
Total gasto com abonos: R${total_abono:.2f}
Valor mínimo pago a {abono_minimo} colaboradores
Maior valor de abono pago: R${maior_abono:.2f}""")

abono_colaboradores()
```

```python
#Exercicio 21
def consumo_combustivel():
  carros = []
  consumos = []
  preco_gasolina = 2.25
  distancia = 1000

  print('Comparativo de Consumo de Combustível\n')
  for i in range(5):
    print(f'Veículo {i + 1}')
    nome = input('Nome: ').strip().title()
    while True:
      try:
        consumo = float(input('Km por litro: '))
        if consumo <= 0:
          print('O consumo deve ser maior que 0. Tente novamente.')
          continue
        break
      except ValueError:
        print('Entrada inválida! Digite um número válido.')
    carros.append(nome)
    consumos.append(consumo)

  print('\nRelatório Final')
  menor_consumo = float('-inf')
  carro_mais_economico = ""

  for i in range(len(carros)):
      litros_consumidos = distancia / consumos[i]
      custo = litros_consumidos * preco_gasolina
      print(f'{i + 1} - {carros[i]:<15} - {consumos[i]:>6.1f} - {litros_consumidos:>6.1f} litros - R$ {custo:>6.2f}')
      if consumos[i] > menor_consumo:
        menor_consumo = consumos[i]
        carro_mais_economico = carros[i]
  print(f'\nO menor consumo é do {carro_mais_economico}.')

consumo_combustivel()
```

```python
#Exercicio 23
def bytes_para_mb(bytes):
  return bytes / (1024 ** 2)

def calcular_percentual(uso, total):
  return (uso / total) * 100

def gerar_relatorio():
  try:
    with open('usuarios.txt', 'r') as arquivo:
      linhas = arquivo.readlines()
  except FileNotFoundError:
    print("Erro: Arquivo 'usuarios.txt' não encontrado.")
    return

  usuarios = []
  espacos = []

  for linha in linhas:
    if linha.strip() == "" or len(linha) < 16:
      continue
      try:
        usuario = linha[:15].strip()
        espaco = int(linha[15:].strip())
        usuarios.append(usuario)
        espacos.append(espaco)
      except ValueError:
        print(f'Erro ao processar a linha: {linha.strip()}')
        continue

  if not usuarios or not espacos:
    print('Nenhum dado válido encontrado no arquivo.')
    return

  espacos_mb = [bytes_para_mb(espaco) for espaco in espacos]
  total_espaco = sum(espacos_mb)
  espaco_media = total_espaco / len(espacos_mb)

  with open('relatorio.txt', 'w') as relatorio:
    relatorio.write('ACME Inc.\t\tUso do espaço em disco pelos usuários\n')
    relatorio.write('=' * 72 + '\n')
    relatorio.write(f"{'Nr.':<4}{'Usuário':<15}{'Espaço utilizado':<20}{'% do uso':<10}\n\n")

    for i, (usuario, espacos_mb) in enumerate (zip(usuario, espacos_mb), start=1):
      percentual = calcular_percentual(espacos_mb, total_espaco)
      relatorio.write(f"{i:<4}{usuario:<15}{espacos_mb:>10.2f} MB{percentual:>17.2f}%\n")

    relatorio.write(f'\nEspaço total ocupado: {total_espaco:.2f} MB\n')
    relatorio.write(f'\nEspaço médio ocupado: {espaco_media:.2f} MB\n')
print("Relatório gerado com sucesso no arquivo 'relatório.txt'.")

gerar_relatorio()
```
