````python
#Exercicio 15
ganha_hora = float(input('Digite o valor ganho por hora R$ '))
hora_trabalhada = int(input('Digite as horas trabalhadas no mês: '))

salario_bruto = ganha_hora * hora_trabalhada
ir =0.11 * salario_bruto
inss = 0.08 * salario_bruto
sindicato = 0.05 * salario_bruto
salario_liquido = salario_bruto - (ir + inss + sindicato)

print(f"""Salário Bruto: R${salario_bruto:.2f}
IR(11%): R${ir:.2f}
INSS(8%): R${inss:.2f}
Sindicato(5%): R${sindicato:.2f}
Salário Liquido: R${salario_liquido:.2f}
""")
````

````python
#Exercicio 17
import math

area = float(input('Digite os metros quadrados da area: '))
litros = area / 6 * 1.1
latas = litros / 18
galoes = litros / 3.6

if latas % 18 != 0:
  latas += 1
  platas = latas * 80
if galoes % 3.6 != 0:
  galoes += 1
  pgaloes = galoes * 25

mlata = int(litros / 18)
mgaloes = int((litros - (mlata * 18)) / 3.6)

if litros - (mlata*18) % 3.6 != 0:
  mgaloes += 1
gtotal = (mlata * 80) + (mgaloes * 25)

print(f"""
Total de litros necessários para pintar: {litros:.0f}L
Apenas latas de 18 litros: {latas:.0f}un, valor total de R${platas:.2f}
Apenas galoes de 3.6 litros: {galoes:.0f}un, valor total de R${pgaloes:.2f}
Mistura: {mlata:.0f} latas + {mgaloes:.0f} galões, valor total de R${gtotal:.2f}""")
````

````python
#Exercicio 18
download = float(input('Digite o tamanho do arquivo em (Mb): '))
wifi = int(input('Digite sua velocidade de internet em (Mbps): '))

vdownload = download / (wifi / 8)
dias = int(vdownload // (24 * 3600))
vdownload %= (24 * 3600)
horas = int(vdownload // 3600)
vdownload %= 3600
minutos = int(vdownload // 60)
segundos = int(vdownload % 60)

print(f"Tempo estimado para o download: {dias} dias, {horas} horas, {minutos} minutos e {segundos} segundos.")
````

````python
# Exercicio 11
salario = float(input('Digite o seu salário R$: '))

if salario <= 280:
  percentual = 20
elif salario <= 700:
  percentual = 15
elif salario <= 1500:
  percentual = 10
else:
  percentual = 5

print('Salário R$', salario)
print('Reajuste:', percentual, '%')

percentual = percentual/100
aumento = salario * percentual
novo_salario = salario + aumento

print('Aumento R$', aumento)
print('Novo Salário R$', novo_salario)
````

````python
#Exercicio 12
def folha_de_pagamento(salario):
  if salario <= 900:
    desconto_ir = 0
  elif salario <= 1500:
    desconto_ir = 5
  elif salario <= 2500:
    desconto_ir = 10
  else:
    desconto_ir = 20

  ir = salario * (desconto_ir / 100)
  inss = salario * 0.1
  fgts = salario * 0.11
  sindicato = salario * 0.03
  total_desconto = ir + inss + sindicato
  salario_liquido = salario - total_desconto

  return f"""Salário Bruto: R${salario:.2f}
  IR ({desconto_ir}%): R${ir:.2f}
  INSS (10%): R${inss:.2f}
  Sindicato (3%): R${sindicato:.2f}
  FGTS (11%): R${fgts:.2f}
  Total de descontos: R${total_desconto:.2f}
  Salário Líquido: R${salario_liquido:.2f}"""

print(folha_de_pagamento(900))
print(folha_de_pagamento(2000))
print(folha_de_pagamento(3000))
````

````python
# Exercicio 15
def triangulo(lado1, lado2, lado3):
  if (lado1 + lado2 > lado3) and (lado2 + lado3 > lado1) and (lado1 +lado3 > lado2):
    print('É um triângulo!')
    if lado1 == lado2 == lado3:
      print('Triângulo Equilátero')
    elif lado1 == lado2 or lado2 == lado3 or lado1 == lado3:
      print('Triângulo Isósceles')
    else:
      print('Triângulo Escaleno')
  else:
    print('Não é um triângulo')

triangulo(1, 3, 2)  # Não é um triângulo
triangulo(12, 3, 6)  # Não é um triângulo
triangulo(3, 3, 3)  # É um triângulo, Triângulo Equilátero
triangulo(5, 5, 8)  # É um triângulo, Triângulo Isósceles
triangulo(3, 4, 5)  # É um triângulo, Triângulo Escaleno
````

````python
#Exercicio 16
a = int(input('Digite o valor de A: '))
b = int(input('Digite o valor de B: '))
c = int(input('Digite o valor de C: '))

if a == 0:
    print('A equação não é do segundo grau.\nPROGRAMA ENCERRADO!')
else:
    delta = b ** 2 - 4 * a * c
    if delta < 0:
        print('A equação não possui raízes reais!\nPROGRAMA ENCERRADO!')
    elif delta == 0:
        print('A equação possui apenas uma raiz real.')
        x1 = -b / (2 * a)
        print('O valor de x1 é:', x1)
    else:
        print('A equação possui duas raízes reais.')
        print('O valor de delta é:', delta)

        x1 = (-b + delta ** 0.5) / (2 * a)
        x2 = (-b - delta ** 0.5) / (2 * a)

        print('O valor de x1 é:', x1)
        print('O valor de x2 é:', x2)
````
